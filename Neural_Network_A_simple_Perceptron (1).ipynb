```python
#1. how do you create simple perceptron for basic binary classification?

import math

class Perceptron:

  def __init__(self, num_inputs, threshold):
    self.weights = [0.0] * num_inputs
    self.bias = 0.0
    self.threshold = threshold

  def predict(self, inputs):
    weigthed_sum = sum(w * x for w , x in zip(self.weights, inputs) ) + self.bias

    output = 1/(1+ math.exp(-weigthed_sum)) # activation function

    return  1 if output > self.threshold else 0


  def train(self, training_data, targets, learning_rate, epochs):
    for epoch in range(epochs):
      for input , target in zip(training_data, targets):
        prediction = self.predict(input) #step 1
        error = target - prediction # step 2
        for i in range(len(self.weights)):

          self.weights[i] += learning_rate * error * input[i]
          self.bias += learning_rate * error


perceptron = Perceptron(num_inputs=2, threshold=.5)

training_data = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]

target = [0,0,0,1]

perceptron.train(training_data=training_data, targets=target, epochs=10, learning_rate=.1)
for input in training_data:
  prediction = perceptron.predict(input)
  print(prediction)


```

```python
# 8 How can you viualize the training process with accuracy and loss curves ?
import keras

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

model = Sequential([
    Flatten(input_shape = (28,28)),
    Dense(units = 132, activation = 'relu'),
    Dense(10, activation = 'softmax')

])

model.compile(optimizer='Adam' , loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model = model.fit(x_train,y_train, epochs=15, batch_size=20)


import matplotlib.pyplot as plt

accuracy = model.history['accuracy']
loss = model.history['loss']

epochs_range = range(1, len(accuracy) +1 )

# visualization for accuracy
plt.subplot(1,2,1)
plt.plot(epochs_range, accuracy, label = 'traning accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Model acuraccy during training')
plt.legend(loc='lower right')
plt.grid(True)


# visualization for loss
plt.subplot(1,2,2)
plt.plot(epochs_range, loss, label = 'traning loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Model loss during training')
plt.legend(loc='upper right')
plt.grid(True)


plt.show()
```

```python
# 10 How can you use gradient clipping in keras to control the gradient size and prevent exploding gradients?
# Ans Exploding gradient happen when the mathematical updates  during back propagation  make worse into massive , unstable number .
# Gradient clipping is the mechanism that maintains that absoulte control , it intercept the gradient before they update the weights
# force them under a predefined ceiling.

import keras
from keras import optimizers
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.models import Sequential

model = Sequential([
    Dense(units =20, input_shape = (10,) ), #layer 1
    BatchNormalization(), #Batch normalization added here
    Activation('relu'),  #Activation fuction follows Batch normalization

    Dense(64, activation='relu'), # layer 2

    Dense(units = 2, activation ='sigmoid')
])


controlled_adam = optimizers.Adam(learning_rate=0.01, clipnorm=1.0) # A clipnorm of 1.0 is the standard starting point

model.compile(
    optimizer =controlled_adam,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# 10 How can you create a custom loss function in keras ?
import tensorflow as tf
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.models import Sequential

def custom_categorical_crossentropy(y_true, y_pred):
  #Clip predictions to prevent from log(0)
  epsilon = 1e-7
  y_pred = tf.clip_by_value(y_pred, epsilon, 1.0 - epsilon)


  #calculate cross entrpy:
  loss = -tf.reduce_sum(y_true * tf.math.log(y_pred), axis =-1)

  # Return the mean loss for the batch
  return tf.reduce_mean(loss)



model = Sequential([
    Dense(units =20, input_shape = (10,) ), #layer 1
    BatchNormalization(), #Batch normalization added here
    Activation('relu'),  #Activation fuction follows Batch normalization

    Dense(64, activation='relu'), # layer 2

    Dense(units = 2, activation ='sigmoid')
])


model.compile(optimizer='adam', loss=custom_categorical_crossentropy)


```
